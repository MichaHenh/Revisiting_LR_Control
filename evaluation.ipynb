```python
from dacbench.logger import Logger, log2dataframe, load_logs
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

```python
# List of JSON file paths
file_paths = glob.glob('results_mnist/cocob_s*/PerformanceTrackingWrapper.jsonl')
# Initialize a list to store the data
data_list = []

# Loop through each file and load the data
for file_path in file_paths:
    with open(file_path, 'r') as file:
        log = load_logs(file_path)
        data = log2dataframe(log)
        data_list.append(data[["episode", "value"]])

# Combine all DataFrames into one
df = pd.concat(data_list, ignore_index=True)

# Group by episode and calculate the mean performance value
df_cocob = df.groupby('episode')['value'].mean().reset_index()
# Rename the columns for clarity
df_cocob.columns = ['episode', 'average_performance']

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(df_cocob['episode'], df_cocob['average_performance'], marker='o', linestyle='-')
plt.xlabel('Episode')
plt.ylabel('Average Performance Value')
plt.title('Average Overall Performance Over Episodes')
plt.grid(True)
plt.show()
```
